```python

fpath_G = 'DL_Project/ROI/b4'
names = []
filesG = [os.path.join(root, name)
            for root, dirs, files in os.walk(fpath_G)
            for name in files
            if (name.endswith((".png", ".PNG")))and ('good' not in name)and ('roi' not in name)]
for root, dirs, files in os.walk(fpath_G):
  for name in files:
    names.append(name)
```

```python
imgG = [] #Undefected cells 
ROI = []  #ROIs of defected cells
for pth in filesG:
  roi = np.empty((4624,3468,3))
  roi2 = np.empty((4624,3468)) 
  img = cv2.imread(pth)/255.
  img = np.array(img)
  for i in range(3468):
    for j in range(4624):
      roi[j][i][:] = img[j][i+3468][:]
  for i in range(3468):
    for j in range(4624):
      if roi[j][i][2] > 0.67 and roi[j][i][2] < 0.83 and roi[j][i][0] > 0.28 and roi[j][i][0] < 0.55:
        roi2[j][i] = 255
      else:
        roi2[j][i] = 0
  ROI.append(roi2)

```

```python
!ls
%cd DataSet/cracked_roi
!ls
for img in ROI:
  for name in names:
    Image.fromarray(img).convert('L').save(name + "_roi.png")
    names.remove(name)
    break
```
